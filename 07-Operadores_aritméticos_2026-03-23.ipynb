## 01 – Obtenha os resultados e retorne o objeto e código em Python dos seguintes cálculos:
a) 2.0 + 3.0 / 5.0

b) (a + 10.2) - 5.0 * 2.0

c) a / b + 33.8

d) a / (b + c)

e) d * 2

```python
a=2.0+3.0/5.0
print(f'a = {a}')

b=(a+10.2)-5.0*2.0
print(f'b = {b}')

c=a/b+33.8
print(f'c = {c}')

d=a/(b+c)
print(f'd = {d}')

e=d*2
print(f'e = {e}')
```
