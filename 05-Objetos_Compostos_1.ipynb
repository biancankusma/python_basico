## 01 – Quais as vantagens da utilização de objetos compostos em detrimento dos objetos simples?

Objetos compostos permitem um melhor aproveitamento da memória e maior controle por armazenarem dados relacionados e seus métodos dentro de um único objeto com uma estrutura organizada, o que também permite realizar algumas operações de forma mais fácil e simplificar a lógica do código que precisa interagir com os objetos.
