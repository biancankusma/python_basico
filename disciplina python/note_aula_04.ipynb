```python
import pandas as pd
import numpy as np
```

```python
simbolo = [
    'IFNG',
    'Il10',
    'Tnf',
    'CSF2'
]
full_name = [
    'interferon gamma',
    'interleukin 10',
    'tumor necrosis factor',
    'colony stimulating factor 2'
]
source = [
    'HGNC:HGNC:5438',
    'RGD:2886',
    'MGI:MGI:104798',
    'HGNC:HGNC:2434'
]
gene_type = [
    'protein coding',
    'protein coding',
    'protein coding',
    'protein coding'
]
organism = [
    'Homo sapiens',
    'Rattus norvegicus',
    'Mus musculus',
    'Homo sapiens'
]
chromosome = [
    12,
    13,
    17,
    5
]
```

```python
#é uma lista de lista, da pra ver pelos colchetes
gene_matriz_lista = [
    simbolo,
    full_name,
    source,
    gene_type,
    organism,
    chromosome
]
gene_matriz_lista
```

```python
dic_1={'simbolo': simbolo, 'full_name': full_name, 'source': source, 'gene_type': gene_type, 'organism': organism, 'chromosome': chromosome}
dic_1
```

```python
gene_array=np.array(gene_matriz_lista)
gene_array
```

```python
gene_array.shape
```

```python
#recuperar o terceiro elemento da primeira linha:
print(gene_array[0,2])
print(gene_array[0][2])
```

```python
#recuperar algumas linhas selecionadas
print(gene_array[0:2])
```

```python
#recuperar linhas e colunas, pensar que é sempre nessa ordem separando por vírgula:
print(gene_array[0:2, 1:3])
```

```python
#os dois pontos sozinho vai pegar tudo e se ta antes ou depois da virgula diz se ta se referindo a linhas (antes) ou colunas (depois)
#se botar [] dentro vai criar uma lista dentro de linha ou coluna, conforme a vírgula
#nesse caso, ta pedindo de todas as linhas, a segunda coluna e a primeira coluna, nessa ordem:
print(gene_array[:,[1,0]])
```
