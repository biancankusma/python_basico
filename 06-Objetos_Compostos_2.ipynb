## 2 – No exercício 9 da última aula, foram instanciadas 5 listas. Instancie um objeto composto multidimensional da classe lista com estes cinco vetores. Recupere a matriz.

```python
list_genes=['GAB2','GRB2','LYN','PTGS1','FGFR3']


lista0=[len(list_genes[0]),list_genes[0][:2],list_genes[0].lower().count('a'),'hu' in list_genes[0]]
lista1=[len(list_genes[1]),list_genes[1][:2],list_genes[1].lower().count('a'),'hu' in list_genes[1]]
lista2=[len(list_genes[2]),list_genes[2][:2],list_genes[2].lower().count('a'),'hu' in list_genes[2]]
lista3=[len(list_genes[3]),list_genes[3][:2],list_genes[3].lower().count('a'),'hu' in list_genes[3]]
lista4=[len(list_genes[4]),list_genes[4][:2],list_genes[4].lower().count('a'),'hu' in list_genes[4]]

matriz_2_9=[lista0,
            lista1,
            lista2,
            lista3,
            lista4]
matriz_2_9
```

## 7 – Converta o objeto instanciado no exercício 10 da aula anterior para uma DataFrame Pandas. Nesta DataFrame, os index devem ser os nomes dos genes e cada coluna deve ter um nome que indique seu conteúdo. Recupere a DataFrame.

```python
import pandas as pd
```

```python
dicio_4_8={list_genes[0]:lista0,
           list_genes[1]:lista1,
           list_genes[2]:lista2,
           list_genes[3]:lista3,
           list_genes[4]:lista4}
dicio_4_8 

#o T ao final da próxima linha faz a transposição da matriz para que cada gene vira uma linha e cada característica vira uma coluna.
df_4_8 = pd.DataFrame(dicio_4_8).T

df_4_8.columns = [
    "num_caract",
    "primeiros_2_caract",
    "quant_a",
    "tem_hu"
]

print(df_4_8)
```
